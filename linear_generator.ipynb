```python
x_array = []
u_array = []

x_0 = 1
a = 17
b = 1
m = 1000

N = 1000
```

```python
for i in range(N):
    if i == 0:
        x_prev = 1
    else:
        x_prev = x_array[i-1]
    x = (a * x_prev + b) % m
    u = x / m

    x_array.append(x)
    u_array.append(u)
```

```python
print(x_array)
print(u_array)
```

```python
Mt = 0.5
M = sum(u_array) / N

q = (M - Mt) / Mt * 100

print(q)
```
